```python
df1 = df[['SKU', 'PRICE']]
df1
# data is in object dtype but mixed format
```

```python
mydf = df1.copy()
```

```python
mydf['PRICE'].apply(type)
# checking for 'dtype'
```

```python
def clean_dollars(price1):
    if isinstance(price1, str):
        return(price1.replace(',', '').replace('?', '').replace('$',''))
    return(price1)

# the function checks for a string input and removes comma(,), ? and $ . 
```

```python
mydf['PRICE'] = mydf['PRICE'].apply(clean_dollars).astype('float')
# convert string to float for further analysis
```

```python
mydf.dtypes
```

```python
mydf['PRICE'].apply(type).value_counts()
```

```python
mydf
```

```python
mydf.groupby(['SKU'], sort=True)
# sorting the price in an ascending order group wise
```

```python
mydf
```

```python
df2 = mydf['SKU'].value_counts().to_frame()
# converting to dataframe so that it can be joined with another dataframe
```

```python
df2.reset_index(inplace=True)
# leveling it in a flat hierarchy
```

```python
df2.rename(columns = {'index': 'SKU', 'SKU': 'COUNT'}, inplace = True)
# checking for count to check for number of SKU occurences in the 'SKU' column 
```

```python
df2
```

```python
data1 = pd.merge(mydf, df2, on='SKU', how='inner')
# implementing inner join
```

```python
data1
```

```python
data1.columns
```

```python
first_min = data1.groupby('SKU')['PRICE'].first().to_frame()
# finding the first minimum value of each SKU group and storing it in a dataframe
```

```python
first_min.reset_index(inplace=True)
first_min
```

```python
first_min.rename(columns = {'PRICE':'PRICE1'}, inplace=True)
first_min
```

```python
second_min = data1.groupby('SKU')['PRICE'].nth(1).to_frame()
second_min.reset_index(inplace=True)
# finding the second minimum value of each SKU group and storing it in the dataframe
```

```python
second_min.rename(columns={'PRICE':'PRICE2'}, inplace=True)
second_min
```

```python
result_df = pd.DataFrame()
# creating an empty dataframe for storing the resulting dataframe
```

```python
result_df['SKU'] = data1.SKU.unique()
```

```python
result_df
```

```python
result_df = pd.merge(result_df, first_min, on='SKU', how='inner')
result_df
```

```python
result_df = pd.merge(result_df, second_min, on='SKU', how='inner')
result_df
```

```python
result_df.rename(columns={'PRICE1':'FIRST_MINIMUM_PRICE', 'PRICE2':'SECOND_MINIMUM_PRICE'}, inplace=True)
result_df
# Price1 and Price2 columns have been renamed to depict their use in the appropriate context
```

```python
result_df.to_csv(path+'lowestPrice.csv', index=False)
# writing the solution to the file named 'lowestPrice.csv'
# Note: The path has been defined above i.e. the output folder
```
